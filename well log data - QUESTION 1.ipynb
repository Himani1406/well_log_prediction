```python
pd.plotting.scatter_matrix(df,alpha = 1, c='black', figsize=(20,20), s=150)
plt.show()
```

```python
k_range = range(2, 4)  # Number of clusters to try
max_iter = 15  # Maximum number of iterations for each trial
tolerance = 0.0001  # Tolerance to stop iteration

for k in k_range:
    # Initialize the KMeans model with the current number of clusters
    kmeans = KMeans(n_clusters=k, max_iter=max_iter, tol=tolerance)
    
    # Fit the model to the data
    kmeans.fit(df)
    
    # Get the cluster assignments for each data point
    labels = kmeans.labels_
    
    # Print the cluster assignments for each data point
    print("Number of clusters: ", k)
    print("Cluster assignments: ", labels)
```

```python
print(kmeans.labels_)
```

```python
scatter = plt.scatter(df['density(g/cc)'],df['Gamma(API)'],
                     c=labels,s=25)

plt.xlabel('porosity')
plt.ylabel('Gamma')
plt.colorbar(scatter)
plt.show()
```
